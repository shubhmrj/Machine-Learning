## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations. 

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import os

def load_dataset(source="iris", is_csv=False):
    """Load dataset from Seaborn or CSV file."""
    try:
        if is_csv:
            if not os.path.exists(source):
                raise FileNotFoundError(f"File '{source}' does not exist.")
            data = pd.read_csv(source)
            print(f"CSV file '{source}' loaded successfully.")
        else:
            data = sns.load_dataset(source)
            print(f"Seaborn dataset '{source}' loaded successfully.")
        return data
    except Exception as e:
        print(f"Error loading dataset: {e}")
        return None

def visualize_dataset(data, x_col=None, y_col=None, hue_col=None, kind="scatter"):
    """Visualize dataset based on the plot type."""
    if data is None:
        print("No data to visualize.")
        return

    if kind == "scatter":
        sns.scatterplot(data=data, x=x_col, y=y_col, hue=hue_col)
    elif kind == "pairplot":
        sns.pairplot(data, hue=hue_col)
    elif kind == "heatmap":
        corr = data.corr(numeric_only=True)
        sns.heatmap(corr, annot=True, cmap="coolwarm")
    else:
        print(f"Visualization kind '{kind}' not supported.")
        return

    plt.title(f"{kind.capitalize()} Plot" if kind != "heatmap" else "Correlation Heatmap")
    plt.show()

def main():
    # Toggle between Seaborn dataset or CSV file
    use_csv = True  # Change to False to use Seaborn datasets
    source = "data.csv" if use_csv else "iris"

    data = load_dataset(source, is_csv=use_csv)

    if data is not None:
        print(data.head())  # Preview

    # Visualizations
    visualize_dataset(data, x_col="sepal_length", y_col="sepal_width", hue_col="species", kind="scatter")  # Adjust columns as per CSV
    visualize_dataset(data, hue_col="species", kind="pairplot")
    visualize_dataset(data, kind="heatmap")

if __name__ == "__main__":
    main()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def load_dataset(name="iris"):
    """Load a dataset by name using Seaborn."""
    try:
        data = sns.load_dataset(name)
        print(f"Dataset '{name}' loaded successfully.")
        return data
    except Exception as e:
        print(f"Error loading dataset '{name}': {e}")
        return None

def visualize_dataset(data, x_col, y_col, hue_col=None, kind="scatter"):
    """Visualize the dataset using Seaborn."""
    if data is None:
        print("No data to visualize.")
        return

    if kind == "scatter":
        sns.scatterplot(data=data, x=x_col, y=y_col, hue=hue_col)
    elif kind == "pairplot":
        sns.pairplot(data, hue=hue_col)
    elif kind == "heatmap":
        corr = data.corr(numeric_only=True)
        sns.heatmap(corr, annot=True, cmap="coolwarm")
    else:
        print(f"Visualization kind '{kind}' not supported.")
        return

    plt.title(f"{kind.capitalize()} Plot of {x_col} vs {y_col}" if kind != "heatmap" else "Correlation Heatmap")
    plt.show()

def main():
    dataset_name = "iris"  # Change as needed
    data = load_dataset(dataset_name)

    # Example visualization
    visualize_dataset(data, x_col="sepal_length", y_col="sepal_width", hue_col="species", kind="scatter")
    visualize_dataset(data, x_col=None, y_col=None, hue_col="species", kind="pairplot")
    visualize_dataset(data, x_col=None, y_col=None, kind="heatmap")

if __name__ == "__main__":
    main()
```
