```python
import gc

class Head():
    def __init__(self,name) :
        self.name=name
        print(f"new var  {self.name} is assigned.")
        
    def __del__(self):
        print(f"this refrence {self.name} is del.")
        
obj=Head("sh")
obj1=Head("ub") 
obj1.ref=obj
obj.ref=obj1   

del obj
del obj1

# Manually Trigerred the garbage collection
gc.collect()
    
```

```python
import tracemalloc
def create_list():
    return [i for i in range(10001)]

def main():
    tracemalloc.start()
    create_list()
    
    snapshot=tracemalloc.take_snapshot()
    top_stats=snapshot.statistics("lineno")
    
    print("[TOP 10]")
    
    for stat in top_stats[:10]:
        print(stat)
```

```python
main()
```
