```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,4,9,16,2,36]
y1=[4,3,8,3,11,9]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y,color="blue")
plt.title("Graph 1")

plt.subplot(2,2,2)
plt.plot(y1,y,color="blue")
plt.title("Graph 2")

plt.subplot(2,2,3)
plt.plot(x,y1,color="blue")
plt.title("Graph 3")

plt.subplot(2,2,4)
plt.plot(y1,x,color="blue")
plt.title("Graph 4")
```

```python
plot_rows=[1,4,8,5,7,11]
plot_column=[10,12,20,13,18,28]


plt.bar(plot_rows,plot_column,color="purple")
plt.xlabel("plot_rows")
plt.ylabel("plot_column")
plt.title("Bar")
plt.show()
```
